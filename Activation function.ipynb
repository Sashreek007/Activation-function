```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid activation function

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
x=-10000
```

```python
print(sig(x))
```

```python
x= np.linspace(-10000,10000,11)
```

```python
Output= sig(x) 
```

```python
plt.xlabel('x')
plt.ylabel('Output')
plt.plot(x, Output)
plt.title('Sigmoid Activation fuction')
plt.show()
```
